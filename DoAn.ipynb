- Hãy chọn đúng Python kernel (môi trường bạn muốn dùng).
- Các cell bên dưới dùng `sys.executable` để đảm bảo lệnh chạy đúng Python của kernel.

```python
from pathlib import Path
import os, sys

def find_project_root(start: Path) -> Path:
    for p in (start, *start.parents):
        if (p / "requirements.txt").exists() and (p / "train.py").exists():
            return p
    raise FileNotFoundError("Không tìm thấy project root (thiếu requirements.txt/train.py).")

root = find_project_root(Path.cwd().resolve())
os.chdir(root)

print("Project root:", root)
print("Python executable:", sys.executable)
```

```python
import subprocess, sys

# Cài thư viện đúng theo kernel hiện tại (an toàn với đường dẫn có khoảng trắng)
_ = subprocess.run([sys.executable, "-m", "pip", "install", "-r", "requirements.txt"], check=True)
print("OK: installed requirements")
```

```python
from pathlib import Path
import json
import matplotlib.pyplot as plt

results_dir = Path("results")
summary_path = results_dir / "summary.json"
print("Results dir:", results_dir.resolve())

if summary_path.exists():
    summary = json.loads(summary_path.read_text(encoding="utf-8"))
    print("Summary rows:", len(summary))
    for row in summary:
        print(row)
else:
    print("Chưa thấy summary.json. Hãy chạy cell train trước.")

# Hiển thị learning curve (nếu có)
curve_candidates = sorted(results_dir.glob("learning_curve_*.png"))
if curve_candidates:
    img_path = curve_candidates[0]
    print("Showing:", img_path.name)
    img = plt.imread(img_path)
    plt.figure(figsize=(8, 4))
    plt.imshow(img)
    plt.axis("off")
    plt.show()
else:
    print("Chưa thấy learning_curve_*.png trong results/.")
```
